# Зависимость цен на нефть и курса рубля
#### График 1. С учетом фактической цены на нефть

```python
import pandas as pd
import pylab
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data= pd.read_csv("d.csv", delimiter=";")
```

```python
data.head()
```

```python
Date = data.Date
Urals = data.Urals
Exchange = data.Exchange
Day = data.day
```

```python
%matplotlib notebook
ax = plt.figure()
three_d = ax.gca(projection='3d')
three_d1 = three_d.scatter(Day, Exchange, Urals, c = Day )
three_d.set_xlabel('Номер наблюдения с января 2015 г.')
three_d.set_ylabel('Курс рубля, руб./долл.')
three_d.set_zlabel('Цена на нефть Urals, долл./барр.')


plt.colorbar(three_d1)
plt.show()
```
